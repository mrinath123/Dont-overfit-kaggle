```python
import pandas as pd
from sklearn import model_selection
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv("train.csv")
```

```python
df['fold'] = -1
```

```python
def cross_fold(data):
    data = data.sample(frac=1).reset_index(drop=True)
    y= data.target.values
    s_folds= model_selection.StratifiedKFold(n_splits=5)
    for f, (t_, v_) in enumerate(s_folds.split(X=data, y=y)):
        data.loc[v_, 'fold'] = f
    return data
```

```python
xx=cross_fold(df)
```

```python
xx.head()
```

```python
for i in range(5):
    print(f'disrtibution of target in fold {i}')
    print(xx[xx['fold']==i].target.value_counts())
```
